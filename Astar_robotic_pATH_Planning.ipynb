```python
def make_refer_matrix(num_nodes,compl_nlist):
    node_list=compl_nlist.copy()
    #cohu_mat=np.zeros((num_nodes+1,num_nodes+2),np.uint16)
    cohu_mat=np.zeros((num_nodes+1,num_nodes+2),np.uint16)
    #print(cohu_mat)
    for row in range (num_nodes+1):
        for col in range (num_nodes+2):
            #print("the euclidean distanceof({},{}) is".format(row,col))
            #print(int(distance.euclidean(node_list[row],node_list[col])))
            cohu_mat[row,col]=(distance.euclidean(node_list[row],node_list[col]))
    #print(cohu_mat)
    return cohu_mat
```
